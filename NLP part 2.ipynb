```python
def extract_words(text,fname):
    stop_words = set(stopwords.words('english'))
    tokenizer = RegexpTokenizer(r'\w+') 
    words = []
    sentences = tokenizer.tokenize(text)
    for s in sentences:
        words += word_tokenize(s)
    filtered_sentence = [w for w in words if not w in stop_words] 
    #w_obj = json.dumps({'id':fname,'val':filtered_sentence})
    return(filtered_sentence)
```

```python
def extract_lemma(words):
    lemma_words =[]
    lemmatizer = WordNetLemmatizer()
    for w in words:
        lemma_words.append(lemmatizer.lemmatize(w))
    #l_obj = json.dumps({'id':words['id'],'val':lemma_words})
    return(lemma_words)
```

```python
def extract_pos(words):
    POS_list = []
    tagged_words = pos_tag(words)
    for word in tagged_words:
        POS_list.append(str(word[0]+'_'+word[1]))
    #p_obj = json.dumps({'id':words['id'],'val':POS_list})
    return POS_list
```

```python
def extract_ne(pswords):
    chunks = []
    for chunk in ne_chunk(pos_tag(pswords)):
        if hasattr(chunk, 'label'):
            chunks.append(' '.join(c[0] for c in chunk)+"_"+chunk.label())
    #ne_obj = json.dumps({'id':pswords['id'],'val':chunks})
    return chunks
```

```python
#tf-idf on the data
from sklearn.feature_extraction.text import TfidfVectorizer
def tf_idf(words_processed):
    word_score ={}
    vectorizer = TfidfVectorizer()
    response = vectorizer.fit_transform(words_processed)
    feature_names = vectorizer.get_feature_names()
    for col in response.nonzero()[1]:
        word_score[feature_names[col]] = response[0, col]
    return word_score
```
